```python
df = pd.read_csv('sum_cust.csv')
#print(df)
costed_event = pd.DataFrame()
costed_event['date'] = df['Customer_Since']

mean_sms_price =  []
mean_call_price = [] 
mean_data_price = []
mean_call_duration = []
mean_data_volume = []
sms_events_per_customer = []
calls_events_per_customer = []
data_events_per_customer = []

N = 5 
for i in range(len(costed_event)):
    #mean_sms_price.append(round(random.uniform(0.2,0.3), N) + i**(random.choice([1,2,3])) * (random.choice([-1,2,3,5]) / 10000))
    mean_sms_price.append(round(random.uniform(0.25,0.3), N) + i * 0.000003)
    mean_call_price.append(round(random.uniform(0.2,0.25), N) + i * 0.00007)
    mean_data_price.append(round(random.uniform(0.06,0.07), N) + i * 0.000005) 
    
    mean_call_duration.append(round(random.uniform(2.9,3.1), N)) 
    mean_data_volume.append(round(random.uniform(65,75), N) + i * 0.01) 
    
    #sms_events_per_customer.append(round(random.uniform(1.9,2.1), N) + i * random.choice([-0.002,0.002])) 
    sms_events_per_customer.append(round(random.uniform(1.4,1.7), N) + i * 0.001) 
    calls_events_per_customer.append(round(random.uniform(2.9,3.1), N) + i * 0.0015) 
    data_events_per_customer.append(round(random.uniform(13,17), N) + i * 0.0017) 
    
    
costed_event['sms_events']   = df['Customer_count'] * sms_events_per_customer
costed_event['calls_events'] = df['Customer_count'] * calls_events_per_customer
costed_event['data_events']  = df['Customer_count'] * data_events_per_customer


costed_event['mean_call_duration'] = mean_call_duration #3 #min
costed_event['mean_data_volume']   = mean_data_volume #20 20 megabyte
 
    
costed_event['mean_sms_cost']  = mean_sms_price   #0.25 
costed_event['mean_call_cost'] = mean_call_price  #0.53 
costed_event['mean_data_cost'] = mean_data_price  #0.002 #per megabyte


costed_event['total_sms_cost']   = costed_event['sms_events']   * costed_event['mean_sms_cost']
costed_event['total_calls_cost'] = costed_event['calls_events'] * costed_event['mean_call_cost']
costed_event['total_data_cost']  = costed_event['data_events']  * costed_event['mean_data_cost']
costed_event['total_cost']  = costed_event['total_sms_cost'] + costed_event['total_calls_cost'] + costed_event['total_data_cost']


costed_event['sms_ARPU']    = costed_event['total_sms_cost']   / df['Customer_count']
costed_event['calls_ARPU']  = costed_event['total_calls_cost'] / df['Customer_count']
costed_event['data_ARPU']   = costed_event['total_data_cost']  / df['Customer_count']
costed_event['total_ARPU']  = costed_event['total_cost']       / df['Customer_count']

costed_event['norm_sms_ARPU']    = costed_event['sms_ARPU']   / costed_event['sms_ARPU'].max()
costed_event['norm_calls_ARPU']  = costed_event['calls_ARPU'] / costed_event['calls_ARPU'].max()
costed_event['norm_data_ARPU']   = costed_event['data_ARPU']  / costed_event['data_ARPU'].max()
costed_event['norm_total_ARPU']  = costed_event['total_ARPU'] / costed_event['total_ARPU'].max()
                          
print(costed_event)
costed_event.to_csv('all_information.csv')
```
