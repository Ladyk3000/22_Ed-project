```python
import pandas as pd 
```

```python
Cities_df = pd.read_csv('Cities_coord.csv').drop(['Unnamed: 0'],axis =1 )
Cities_df.head()
```

```python
customers_df = pd.read_csv('Customers.csv').drop(['Unnamed: 0'],axis =1 )
customers_df.head()
```

```python
new_city_df = pd.DataFrame()

new_city_df['Name'] = pd.unique(Cities_df['Name'])
series = Cities_df.groupby('Name')['Population'].sum()
new_city_df['Population'] = series.tolist()
#print(new_city_df)

new_city_df.loc[new_city_df.Population < 10000, 'type'] = 'village'
new_city_df.loc[(new_city_df.Population > 10000) & (new_city_df.Population < 250000), 'type'] = 'town'
new_city_df.loc[new_city_df.Population > 250000, 'type'] = 'city'
print(new_city_df)

cities = new_city_df.copy()

from opencage.geocoder import OpenCageGeocode
import pandas as pd

key = 'beab1e05a03e4a03a59a535af5d91ab9'
geocoder = OpenCageGeocode(key)
list_Coord = []  

uncoord = ['Ash-Shimāiyah',"Muḥā'yil","Muḥā'yil","R'ās Tanūrah","Ad-Dā'yir",'Khulaīṣ','Al-Madīnah al-Munuwarah']
cities = cities[cities.Name.isin(uncoord) == False]

for city in cities['Name']:
	query = str(city)
	#query = 'Ash-Shimāiyah'
	results = geocoder.geocode(query) 
	#print(results)
	if len(results) != 0:
		list_Coord.append('['+ str(results[0]['geometry']['lat']) +','+ str(results[0]['geometry']['lng']) +']')  
	else:
		print(city)

# create new columns from lists
cities['Coords'] = list_Coord 
cities.to_csv('Cities_coord.csv')
print(cities)
```

```python
custs = customers_df.groupby('Customer_City')['Customer_ID'].count()
#print(custs)

cities = cities.sort_values('Name')
cities['Cusomers_Count'] = custs.tolist()
cities['Cusomers_Percent'] = round(cities['Cusomers_Count'] / cities['Population'], 6)
#print(cities.sort_values('Cusomers_Percent'))
#cities.to_csv('Cities_NEW.csv')
print(cities)
```
